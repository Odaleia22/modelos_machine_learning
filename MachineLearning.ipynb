**mean_absolute_error**: O MAE é uma métrica de avaliação usada em modelos de regressão para medir a precisão das previsões. Ele calcula a média das diferenças absolutas entre os valores previstos e os valores reais, sem considerar a direção do erro (se é positivo ou negativo).

Em termos simples, o MAE fornece uma ideia de quão longe, em média, as previsões estão dos valores reais. Um MAE menor indica previsões mais precisas1.

**Acurácia:**
Acurácia mede a proporção de todas as previsões corretas (tanto positivas quanto negativas) em relação ao total de previsões. Em outras palavras, é uma medida de quão bem o modelo está performando no geral.


**Precisão (Precision)**
Precisão mede a proporção de verdadeiros positivos (TP) entre todas as previsões positivas feitas pelo modelo (TP + FP). Isso indica quantas das previsões positivas do modelo estavam corretas.


**Recall (Sensibilidade):**
Recall mede a proporção de verdadeiros positivos (TP) em relação ao total de casos positivos reais na amostra (TP + FN). Isso mostra quantos dos casos positivos reais foram corretamente identificados pelo modelo. A fórmula é:
Recall=Verdadeiros Positivos (TP)+Falsos Negativos (FN)Verdadeiros Positivos (TP)​

```python
# prompt: explique cada linha de código deste notebook em portugues. Cada codigo e explicação em sua linha

# Importa as bibliotecas necessárias
import pandas as pd
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_absolute_percentage_error
from sklearn.linear_model import LogisticRegression
from mlxtend.plotting import plot_confusion_matrix
from sklearn.metrics import confusion_matrix,classification_report

# Carrega o dataset
df = pd.read_csv('USA_Housing.csv')

# Remove a coluna 'Address'
df_drops = df.drop('Address', axis=1)

# Calcula a correlação entre as colunas
df_drops.corr()

# Seleciona as colunas independentes
x = df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms', 'Avg. Area Number of Bedrooms', 'Area Population']]

# Seleciona a coluna dependente
y = df['Price']

# Divide os dados em treino e teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y)

# Cria o modelo de regressão linear
modelo = LinearRegression()

# Treina o modelo
modelo.fit(x_treino, y_treino)

# Faz previsões
y_previsto = modelo.predict(x_teste)

# Calcula o erro médio absoluto
mean_absolute_error(y_previsto, y_teste)

# Calcula o erro médio absoluto percentual
mean_absolute_percentage_error(y_previsto, y_teste)

# Carrega o dataset de câncer de mama
df = pd.read_csv('breast-cancer.csv')

# Renomeia as colunas
renomeadas = df.rename(columns={
    'id': 'id',
    'diagnosis': 'diagnóstico',
    'radius_mean': 'raio_médio',
    'texture_mean': 'textura_média',
    'perimeter_mean': 'perímetro_médio',
    'area_mean': 'área_média',
    'smoothness_mean': 'suavidade_média',
    'compactness_mean': 'compacidade_média',
    'concavity_mean': 'concavidade_média',
    'concave_points_mean': 'pontos_côncavos_médios',
    'symmetry_mean': 'simetria_média',
    'fractal_dimension_mean': 'dimensão_fractal_média',
    'radius_se': 'raio_se',
    'texture_se': 'textura_se',
    'perimeter_se': 'perímetro_se',
    'area_se': 'área_se',
    'smoothness_se': 'suavidade_se',
    'compactness_se': 'compacidade_se',
    'concavity_se': 'concavidade_se',
    'concave_points_se': 'pontos_côncavos_se',
    'symmetry_se': 'simetria_se',
    'fractal_dimension_se': 'dimensão_fractal_se',
    'radius_worst': 'pior_raio',
    'texture_worst': 'pior_textura',
    'perimeter_worst': 'pior_perímetro',
    'area_worst': 'pior_área',
    'smoothness_worst': 'pior_suavidade',
    'compactness_worst': 'pior_compacidade',
    'concavity_worst': 'pior_concavidade',
    'concave_points_worst': 'piores_pontos_côncavos',
    'symmetry_worst': 'pior_simetria',
    'fractal_dimension_worst': 'pior_dimensão_fractal'
})

# Seleciona as colunas independentes
x = renomeadas.drop(columns=['id', 'diagnóstico'], axis=1)

# Seleciona a coluna dependente
y = renomeadas['diagnóstico']

# Cria o modelo de regressão logística
modelo = LogisticRegression(max_iter=10000)

# Divide os dados em treino e teste
x_treino, x_teste, y_treino, y_teste = train_test_split(x, y)

# Treina o modelo
modelo.fit(x_treino, y_treino)

# Faz previsões
y_previsto = modelo.predict(x_teste)

# Calcula a matriz de confusão
cm = confusion_matrix(y_teste, y_previsto)

# Plota a matriz de confusão
plot_confusion_matrix(cm)

# Imprime o relatório de classificação
print(classification_report(y_teste, y_previsto))
```
